```python
#Create the table headers  for inputs, weight calculations
hebbs.cell(row = 1, column = 1).value = 'Letter'

#x1-x70 | B = 1
for i in range(1, 71):
    hebbs.cell(row = 1, column = 1 + i).value = 'x' + str(i)    
hebbs.cell(row = 1, column = 72).value = 'B = 1'

#t
hebbs.cell(row = 1, column = 73).value = 't'

#Dw1-Dw70 | Db
for i in range(1, 71):
    hebbs.cell(row = 1, column = 73 + i).value = 'Dw' + str(i)
hebbs.cell(row = 1, column = 144).value = 'Db'

#w1-w70 | b
for i in range(1, 71):
    hebbs.cell(row = 1, column = 144 + i).value = 'w' + str(i)
hebbs.cell(row = 1, column = 215).value = 'b'
```

```python
# Converting graphical data to biploar data
for i in range(0, len(data)):
    columnNo = 2
    for row in range(0, 10):
        for col in range(0, 7):
            if data[i][row][col] == 'x':
                hebbs.cell(row = 2 + i, column = columnNo).value = 1
            else:
                hebbs.cell(row = 2 + i, column = columnNo).value = -1
            columnNo +=1
    hebbs.cell(row = 1 + i, column = 72).value = 1
wb.save("Character Classification Hebbs.xlsx")

#Enter the targets manually in t column (column number 73)
#######################################################################################################
#######################################################################################################
```

```python
df = pd.read_excel("Character Classification Hebbs.xlsx")
```

```python
df
```

```python
def calculate_weights(row): 
    # CALCULATE CHANGE IN WEIGHTS -> Dw(i) = xi * t
        for i in range(1, 71):
            df[str('Dw' + str(i))][row] = df[str('x' + str(i))][row] * df['t'][row] 
    #     Db = t
        df['Db'][row] =df['t'][row]

    #######################################
    # UPDATE WEIGHTS -> w(i) = w(i-1) + Dwi
        if row == 0:
            #Assuming initial weights to be 0
            for i in range(1, 71):
                df[str('w' + str(i))][row] = df[str('Dw' + str(i))][row]
            df['b'][row] = df['Db'][row] 
        else:
            for i in range(1, 71):
                df[str('w' + str(i))][row] = df[str('w'+ str(i))][row - 1] + df[str('Dw' + str(i))][row]
            df['b'][row] = df['b'][row - 1] + df['Db'][row]
```

```python
for row in range(0, 8):
    calculate_weights(row)
```

```python
df.to_excel('trained_network_hebbs.xlsx')
```
